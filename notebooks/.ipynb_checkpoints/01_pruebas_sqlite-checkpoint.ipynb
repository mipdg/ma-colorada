```python
import sqlite3
from pathlib import Path

db_path = Path("data/ma_colorada.sqlite")
db_path.parent.mkdir(exist_ok=True)

conn = sqlite3.connect(db_path)

conn.execute("""
CREATE TABLE IF NOT EXISTS pruebas (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nombre TEXT,
    valor REAL
)
""")

conn.execute(
    "INSERT INTO pruebas (nombre, valor) VALUES (?, ?)",
    ("test", 123.45)
)

conn.commit()

rows = conn.execute("SELECT * FROM pruebas").fetchall()
rows
```

```python
import pandas as pd

df = pd.read_sql_query("SELECT * FROM pruebas", conn)
df
```
